```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
stem = "/Users/bandari/Documents/git.repos/rrlyrae_metallicity/ew_products/"

# empirical Stripe 82 EW values
empirical_sdss_ews_file = "all_sdss_spectra_20211118_restacked_ew_info_good_only_w_net_balmer_errors.csv"

# empirical McDonald EW values
empirical_mcd_ews_file = "../bin/retrieved_vals_20211103.csv"

# synthetic 
synthetic_ews_file = "synthetic_ews_test_20211208.csv"
```

```python
df_empirical_sdss = pd.read_csv(stem + empirical_sdss_ews_file)
df_empirical_mcd = pd.read_csv(stem + empirical_mcd_ews_file)
df_synthetic = pd.read_csv(stem + synthetic_ews_file)
```

```python
bins_choice = np.linspace(0, 20, 100)

#plt.hist(df_empirical_sdss["EW_Hbeta"], bins=bins_choice, alpha=0.3, density=True, label='empirical (SDSS)')
plt.hist(df_empirical_mcd["EW_Hbeta"], bins=bins_choice, alpha=0.5, density=True, label='empirical (McD)')
plt.hist(df_synthetic["EW_Hbeta"], bins=bins_choice, alpha=0.5, density=True, label='synthetic')
plt.xlim([0,20])
plt.title("Hbeta EWs")
plt.xlabel("EW (angstr)")
plt.ylabel("Normalized histograms")
plt.legend()
plt.savefig("junk_hbeta_empirical_synthetic_comparison.pdf")
```

```python
plt.clf()
plt.scatter(df_empirical_sdss["EW_Balmer"],df_empirical_sdss["err_EW_Balmer_based_Robo"],s=2, label="empirical (SDSS)")
plt.scatter(df_empirical_mcd["EW_Balmer"],df_empirical_mcd["err_EW_Balmer_based_Robo"],s=2, label="empirical (McDonald)")
plt.scatter(df_synthetic["EW_Balmer"],df_synthetic["err_EW_Balmer_based_Robo"],s=2, label="synthetic")
plt.xlim([0,22])
plt.ylim([0,25])
plt.title("Robospect errors")
plt.xlabel("Balmer EW (angstr)")
plt.ylabel("Error in Balmer EW, Robospect (angstr)")
plt.legend()
plt.savefig("junk.pdf")
```

```python
plt.clf()
plt.scatter(df_empirical_sdss["EW_Balmer"],np.divide(df_empirical_sdss["err_EW_Balmer_based_Robo"],df_empirical_sdss["EW_Balmer"]),s=2, label="empirical (SDSS)")
plt.scatter(df_empirical_mcd["EW_Balmer"],np.divide(df_empirical_mcd["err_EW_Balmer_based_Robo"],df_empirical_mcd["EW_Balmer"]),s=2, label="empirical (McDonald)")
plt.scatter(df_synthetic["EW_Balmer"],np.divide(df_synthetic["err_EW_Balmer_based_Robo"],df_synthetic["EW_Balmer"]),s=2, label="synthetic")
plt.xlim([0,22])
plt.ylim([0,2])
plt.title("Robospect RELATIVE errors")
plt.xlabel("Balmer EW (angstr)")
plt.ylabel("(error / amplitude) in Balmer EW, Robospect (angstr)")
plt.legend()
plt.savefig("junk.pdf")
#plt.show()
```

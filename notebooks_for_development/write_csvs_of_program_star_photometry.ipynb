```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
skiprows = 0
file_string = "vxher_moore_aavso"
write_out_file_name = file_string + "_standard.csv"
df = pd.read_csv("maxima_macadam/"+file_string+".csv",skiprows=skiprows)
#df = pd.read_csv("maxima_macadam/"+file_string+".csv")
```

```python
for key in df.keys():
    print([key])
```

```python
# MacAdam stars

plt.scatter(df['Julian_Day'],df['V_Mag'])
plt.show()
```

```python
# MacAdam stars

plt.scatter(df['Julian_Day'],df['V_Mag']-df['C5'])
plt.show()
```

```python
# Moore stars

#plt.scatter(df["J.D.-2400000"],df['AAVSO Magnitude (Jul 2013)'])
#plt.scatter(df["J.D.-2400000"],df["rel_flux_T1"])
plt.scatter(df["J.D.-2400000"],df["rel_flux_T1"]/df["rel_flux_C4"])
plt.show()
```

```python
# MacAdam stars

#df_new = df[["Julian_Day","V_Mag"]]

df_new = df[['Julian_Day']]
df_new = df_new.rename(columns={'Julian_Day': "JD"})
df_new["del_mag"] = df['V_Mag']#-df['      C3']
'''
df_new = df[['JD']]
#df_new = df_new.rename(columns={'1': "V_Mag"})
df_new["del_mag"] = df['1']
'''
```

```python
df_new
```

```python
# Moore stars

df_new = df[["J.D.-2400000"]]
df_new["J.D.-2400000"] = np.add(2400000,df_new["J.D.-2400000"])
df_new["del_mag"] = -2.5*np.log10(df["rel_flux_T1"]/df["rel_flux_C4"])
df_new = df_new.rename(columns={"J.D.-2400000": "JD"})
```

```python
plt.scatter(df_new["JD"],df_new["del_mag"])
plt.show()
```

```python
df_new
```

```python
# write

df_new.to_csv(write_out_file_name,index=False)
print(write_out_file_name)
```
